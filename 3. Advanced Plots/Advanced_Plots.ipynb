# **Advanced Plots Using Matplotlib**

## 1. Box Plot
**Objective:** Display the distribution of data based on quartiles and identify outliers.

**Explanation:** A box plot (or box-and-whisker plot) shows the median, quartiles, and potential outliers in a dataset. It provides a visual summary of the data's spread and central tendency.

* `plt.boxplot()`: Creates a box plot.
* `notch=True`: Adds a notch to show the confidence interval around the median.
* `vert=False`: Makes the box plot horizontal.
* `patch_artist=True`: Fills the boxes with color.
* `boxprops`, `whiskerprops`, and `medianprops`: Customize the appearance of the box, whiskers, and median line.
* `plt.grid()`: Adds a grid for better readability.

### **Code Example:**

```python
import matplotlib.pyplot as plt

# Data for plotting
data = [12, 15, 14, 10, 8, 19, 21, 24, 29, 22, 17, 23, 18, 26, 30]

plt.boxplot(data, notch=True, vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'),
            medianprops=dict(color='red'))

plt.xlabel('Value')
plt.title('Box Plot of Data')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## 2. Heatmap
**Objective:** Visualize matrix-style data where values are represented by colors.

**Explanation:** Heatmaps display data in a 2D grid where the color intensity represents the value. They are useful for understanding the density and patterns in matrix data.

* `plt.imshow()`: Displays the data as an image.
* `cmap='viridis'`: Specifies the colormap (color scheme) used for the heatmap.
* `interpolation='nearest'`: Ensures the heatmap cells are not smoothed.
* `plt.colorbar()`: Adds a color bar to show the intensity scale.

### **Code Example:**

```python
import numpy as np

# Data for plotting
data = np.random.rand(10, 10)  # Generating a 10x10 matrix of random values

plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Intensity')
plt.title('Heatmap of Random Data')
plt.show()
```

## 3. Contour Plot
**Objective:** Display three-dimensional data in two dimensions using contour lines.

**Explanation:** Contour plots represent three-dimensional data in two dimensions. They use contour lines to show regions with equal values.

* `plt.contour()`: Creates a contour plot.
* `levels=10`: Specifies the number of contour levels.
* `cmap='plasma'`: Specifies the colormap used for the contour lines.
* `plt.colorbar()`: Adds a color bar to indicate the value ranges.

### **Code Example:**

```python
# Data for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Z values based on the sine function

plt.contour(X, Y, Z, levels=10, cmap='plasma')
plt.colorbar(label='Value')
plt.title('Contour Plot of Sin Function')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## 4. 3D Plot
**Objective:** Visualize three-dimensional data.

**Explanation:** 3D plots allow you to visualize data in three dimensions, showing relationships between three variables.

* `fig.add_subplot(111, projection='3d')`: Creates a 3D subplot.
* `ax.plot_surface()`: Plots a 3D surface plot.
* `cmap='inferno'`: Specifies the colormap used for the surface plot.
* `ax.set_xlabel()`, `ax.set_ylabel()`, `ax.set_zlabel()`, and `ax.set_title()`: Add labels and title to the plot.

### **Code Example:**

```python
from mpl_toolkits.mplot3d import Axes3D

# Data for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='inferno')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of Sin Function')

plt.show()
```
