# **Matplotlib Visualization Tips Customization and Subplots**

## **Customization and Aesthetics**

## Titles and Labels
**Objective:** Make your plots more informative by customizing titles, axis labels, and legends.

**Explanation:** Titles and labels help viewers understand what the plot represents. Legends differentiate multiple data series.

* `plt.xlabel()`, `plt.ylabel()`, and `plt.title()`: Set the labels and title for the plot.
* `plt.legend()`: Adds a legend to identify the data series.
* `plt.grid(True)`: Adds a grid for better readability.

### **Code Example:**

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='-', color='b', label='Prime Numbers')
plt.xlabel('X-axis: Index')
plt.ylabel('Y-axis: Values')
plt.title('Plot of Prime Numbers')
plt.legend()
plt.grid(True)

plt.savefig('Titles and Labels.png')
plt.show()
```

## **Styles and Colors**
**Objective:** Enhance the visual appeal by changing plot styles and colors.

**Explanation:** Customizing styles and colors helps in making plots more visually distinct and appealing.

* `plt.style.use()`: Applies a predefined style to the plot.
* Customizations include marker, linestyle, color, and linewidth.

### **Code Example:**

```python
plt.style.use('ggplot')  # Use a predefined style

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='--', color='red', linewidth=2)
plt.xlabel('X-axis: Index')
plt.ylabel('Y-axis: Values')
plt.title('Styled Plot with Custom Colors')
plt.grid(True)

plt.savefig('Styles and Colour.png')
plt.show()
```

## **Annotations**
**Objective:** Add text and markers to highlight specific data points.

**Explanation:** Annotations are used to emphasize important data points or provide additional information.

* `plt.annotate()`: Adds text annotations to the plot. xy is the point to annotate, and xytext is the position of the text.

### **Code Example:**

```python
# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.xlabel('X-axis: Index')
plt.ylabel('Y-axis: Values')
plt.title('Plot with Annotations')

# Add annotation
plt.annotate('Prime Number 7', xy=(4, 7), xytext=(3, 8),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.grid(True)
plt.savefig('Anotations.png')
plt.show()
```

## **Grid and Ticks**
**Objective:** Customize grid lines and tick marks for better readability.

**Explanation:** Adjusting grid lines and ticks helps in making the plot clearer and more readable.

* `plt.grid()`: Customizes the grid's appearance.
* `plt.xticks()` and `plt.yticks()`: Customize the ticks and their labels on the x and y axes.

### **Code Example:**

```python
# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.xlabel('X-axis: Index')
plt.ylabel('Y-axis: Values')
plt.title('Plot with Custom Grid and Ticks')

# Customize grid and ticks
plt.grid(True, linestyle='--', color='grey', alpha=0.7)
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['One', 'Two', 'Three', 'Four', 'Five'])
plt.yticks(ticks=[2, 4, 6, 8, 10, 12], labels=['Two', 'Four', 'Six', 'Eight', 'Ten', 'Twelve'])

plt.savefig('Grid and Ticks.png')
plt.show()
```

## **Subplots and Multiple Plots**

## Subplots
**Objective**: Display multiple plots in a single figure using subplots.

**Explanation**: Subplots are used to create multiple plots in a single figure for comparative analysis.

`fig, axs = plt.subplots()`: Creates a grid of subplots.
`axs[i, j]`: Accesses each subplot to plot different data.
`plt.tight_layout()`: Adjusts the layout to prevent overlapping of subplots.

### **Code Example:**

```python
import matplotlib.pyplot as plt

# Create a figure and set of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Data for plotting
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Plot in the first subplot
axs[0, 0].plot(x, y1, 'r--')
axs[0, 0].set_title('Plot 1')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Plot in the second subplot
axs[0, 1].bar(x, y2, color='blue')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

# Plot in the third subplot
axs[1, 0].scatter(x, y1, color='green')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

# Plot in the fourth subplot
axs[1, 1].hist(y2, bins=5, color='purple')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Bins')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap

plt.savefig('Subplots.png')
plt.show()
```

## **Twin Axes**
**Objective:** Create plots with different y-axes on the same x-axis.

**Explanation:** Twin axes are used when you want to plot two different y-axes sharing the same x-axis.

* `ax1.twinx()`: Creates a new y-axis on the right side of the plot.
* `ax1 and ax2`: Plot different datasets with different y-axes.

### **Code Example:**

```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [50, 60, 70, 80, 90]

fig, ax1 = plt.subplots()

# Plot the first y-axis data
ax1.plot(x, y1, 'b-', label='Y1 Data')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y1 Data', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r--', label='Y2 Data')
ax2.set_ylabel('Y2 Data', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Add titles and labels
plt.title('Plot with Twin Axes')
fig.tight_layout()

plt.savefig('Twin Axes.png')
plt.show()
```

## **Figure Size and Layout**
**Objective:** Adjust the figure size and layout for better presentation.

**Explanation:** Customizing figure size and layout helps in fitting the plots nicely into the figure window.

* `plt.figure(figsize=(12, 6))`: Sets the size of the figure.
* `plt.tight_layout()`: Adjusts the layout to prevent overlap and ensure everything fits well.

### **Code Example:**

```python
# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(12, 6))  # Set the figure size

plt.plot(x, y, marker='o', linestyle='-', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Figure Size and Layout')
plt.grid(True)
plt.tight_layout()  # Adjust layout to fit labels and titles

plt.savefig('Figure Size and Layout.png')
plt.show()
```
